```python
#Aplicando as médias móveis
datasetGafisa = pd.DataFrame(datasetGafisa[['Close','Date']])
datasetGafisa.set_index('Date', inplace=True)
datasetGafisa['MM_8'] = datasetGafisa['Close'].rolling(8).mean().shift() 
datasetGafisa['MM_21'] = datasetGafisa['Close'].rolling(21).mean().shift()
```

```python
#Previsão do preço dos próximos 8 dias pelas médias móveis
import numpy as np
```

```python
datasetGafisa['MM_forward_8'] = np.NaN
```

```python
def make_window(window_size, start_point):
    return [start_point+x for x in range(window_size)]
```

```python
limit = 252
window_size = 8
forward_days = 8 # > 1, 
for index in range(window_size, len(datasetGafisa), forward_days):
    for i in range(0,forward_days):
        if index+i >= len(datasetGafisa):
            break
        window_close = make_window(window_size-i, index+i-window_size)
        #print(window_close)
        window_MA = make_window(i, index)
        #print(window_MA)
        mean = pd.concat([ datasetGafisa['Close'].iloc[window_close], datasetGafisa['MM_forward_8'].iloc[window_MA] ]).mean(axis=0)
        datasetGafisa.iat[index+i, datasetGafisa.columns.get_loc('MM_forward_8')] = mean
```

```python
plt.figure(figsize = (15,10))

size = len(datasetGafisa)-limit - (len(datasetGafisa)-limit)%forward_days


for index in range(size, len(datasetGafisa), forward_days):
    plt.plot(datasetGafisa['MM_forward_8'][index:index+forward_days], color='r')
    

plt.plot(datasetGafisa['MM_forward_8'][index:index+forward_days], color='r', label='Preço Previsto' )   
plt.plot(datasetGafisa['Close'][-limit:], color='b', label='Preço de Fechamento' )
plt.legend(title = "Gafisa")
plt.show()


```

```python
#Exponential Moving Avarege
```

```python
datasetGafisa['EMA_DayByDay'] = np.NaN
```

```python
window_size = 8
multiplier = 2 / (window_size + 1) 
#for the first value we use thw SMA
window = make_window(window_size, 0)
EMA_anterior = datasetGafisa['Close'].iloc[window].mean(axis=0)
datasetGafisa.iat[window_size, datasetGafisa.columns.get_loc('EMA_DayByDay')] = EMA_anterior

for index in range(window_size+1, len(datasetGafisa)):
    window = make_window(window_size, index-window_size)
    EMA = (datasetGafisa['Close'].iloc[index-1] - EMA_anterior) * multiplier + EMA_anterior
    datasetGafisa.iat[index, datasetGafisa.columns.get_loc('EMA_DayByDay')] = EMA
    EMA_anterior = EMA
```

```python
plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(datasetGafisa['Close'], label='Preço de Fechamento')
plt.plot(datasetGafisa['EMA_DayByDay'], label='Preço Previsto')
plt.legend(loc=2)
plt.show()
```
